lets train the model using linear regression algorithm

```python

x=df.drop(columns='SALES_PRICE')
x
```

```python
y=df[['SALES_PRICE']]
y
```

we have created a features and target variables so let split the data into training and  testing using train test split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)

model=LinearRegression()
model.fit(x_train,y_train)
model_pred=model.predict(x_test)
```

so we have trained the model so lets check the metrics 

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, model_pred)
print(f"Mean Absolute Error (MAE): {mae}")
mse = mean_squared_error(y_test, model_pred)
print(f"Mean Squared Error (MSE): {mse}")
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")
r2 = r2_score(y_test, model_pred)
print(f"R-squared (R²): {r2}")
```

so we have successfully predicted the values using linear regression model
and evaluated the metrics from the above metrics we can see that r2 score is 0.95 which means
close to 1 so the model fits the data well and predicted the values
