so we gonna start the sentiment analysis process

to do that we need to import the nlkt library which means natural language processing kit from that we gonna downlaod the lexicon parameter this lexicon paramter will set the polarity score for each review and categorize into positive negative and neutral

```python
import nltk
nltk.download('vader_lexicon')
sent=SentimentIntensityAnalyzer()

df['Positive']=[sent.polarity_scores(i)['pos'] for i in  df['Review']]
df['Negative']=[sent.polarity_scores(i)['neg'] for i in  df['Review']]
df['Neutral']=[sent.polarity_scores(i)['neu'] for i in  df['Review']]

df

```
